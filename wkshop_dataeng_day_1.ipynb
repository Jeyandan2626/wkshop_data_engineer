```python
bu = df[df['department'] == 'Computer Science']

print(len(bu))

print(bu[['name', 'math_score', 'programming_score', 'attendance_percentage']])
```

```python
bu = df[df['attendance_percentage'] >= 90]

print(len(bu))
print(bu[['name', 'department', 'attendance_percentage']])
```

```python
bu = df[(df['department'] == 'Computer Science') &
        (df['attendance_percentage'] >= 90)]

print(len(bu))
print(bu[['name', 'math_score', 'programming_score', 'attendance_percentage']])
```

```python
top = df[df['programming_score'] == df['programming_score'].max()]

print(top[['name', 'department', 'programming_score']])
```

```python
max_score = df.groupby('department')[[
    'math_score',
    'science_score',
    'english_score',
    'programming_score'
]].max()

print(max_score)
```

```python
avg_score = df.groupby('department')[[
    'math_score',
    'science_score',
    'english_score',
    'programming_score'
]].mean()

print(avg_score)
```

```python
print("Maximum Scores:")
print(df[['math_score','science_score',
          'english_score','programming_score']].max())

print("\nMinimum Scores:")
print(df[['math_score','science_score',
          'english_score','programming_score']].min())
```

```python
print("Maximum Math Score :", df['math_score'].max())
print("Minimum Math Score :", df['math_score'].min())
print("Average Math Score :", df['math_score'].mean())
print("Average of programming score:", df['programming_score'].mean())
print("average attendance:", df['attendance_percentage'].mean())
```
